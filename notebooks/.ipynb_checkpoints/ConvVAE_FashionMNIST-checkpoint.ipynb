# Convolutional Variational Autoencoder (ConvVAE) on Fashion-MNIST

Bu notebook, ConvVAE projesinin hızlı prototipini ve görselleştirmelerini içerir.

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
import sys
sys.path.append("../src")
```

```python
import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
```

```python
transform = transforms.Compose([transforms.ToTensor()])
train_dataset = datasets.FashionMNIST("data/fashion-mnist", train=True, download=True, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
import matplotlib.pyplot as plt
images, labels = next(iter(train_loader))
fig, axes = plt.subplots(3, 3, figsize=(6, 6))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(images[i].squeeze(), cmap="gray")
    ax.set_title(f"Label: {labels[i].item()}")
    ax.axis("off")
plt.tight_layout()
```

```python
from model import ConvVAE
from utils import visualize_reconstruction, sample_images
import os
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = ConvVAE(latent_dim=32).to(device)
ckpt_path = "../checkpoints/convvae_fashionmnist.pth"
if os.path.exists(ckpt_path):
    model.load_state_dict(torch.load(ckpt_path, map_location=device))
    print("ConvVAE checkpoint yüklendi.")
model.eval()
```

```python
# Rekonstrüksiyonları görselleştir
visualize_reconstruction(model, train_loader, device, num_images=8)
```

```python
# Yeni örnekler üret
sample_images(model, device, num_samples=16, latent_dim=32)
```
